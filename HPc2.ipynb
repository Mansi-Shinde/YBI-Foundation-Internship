Bubble sort

The main focus of this code is the bubble function, which implements the Bubble Sort algorithm. Here's how the code works:

The bubble function takes an array a and the number of elements n as input.

The outer loop runs n times, representing the number of passes in the Bubble Sort algorithm.

The variable first is used to alternate the starting point for each pass. This is done to optimize the parallelization using OpenMP.

The #pragma omp parallel for directive is used to parallelize the inner loop, which performs the comparison and swapping of adjacent elements. The shared(a, first) clause ensures that the array a and the first variable are shared among the threads.

Within the parallel region, the loop iterations are distributed among the available threads for concurrent execution.

The swap function is used to swap two elements in the array.

In the main function:

The user is prompted to enter the total number of elements and then the individual elements of the array.

The bubble function is called with the array and the number of elements.

Finally, the sorted array is printed to the console.

OpenMP is used in this code to potentially parallelize the sorting process, allowing multiple threads to work on different portions of the array simultaneously. This can improve the sorting performance for large arrays. To compile and run this code with OpenMP support, you need to have a compiler that supports OpenMP and enable the OpenMP flag during compilation (e.g., -fopenmp for GCC).

```python
#include<iostream>
#include<stdlib.h>
#include<omp.h>
using namespace std;

void bubble(int *, int);
void swap(int &, int &);


void bubble(int *a, int n)
{
    for(  int i = 0;  i < n;  i++ )
     {  	 
   	 int first = i % 2; 	 

   	 #pragma omp parallel for shared(a,first)
   	 for(  int j = first;  j < n-1;  j += 2  )
   	   {  	 
   		 if(  a[ j ]  >  a[ j+1 ]  )
   		  {  	 
     			 swap(  a[ j ],  a[ j+1 ]  );
   		  }  	 
   		   }  	 
     }
}


void swap(int &a, int &b)
{

    int test;
    test=a;
    a=b;
    b=test;

}

int main()
{

    int *a,n;
    cout<<"\n enter total no of elements=>";
    cin>>n;
    a=new int[n];
    cout<<"\n enter elements=>";
    for(int i=0;i<n;i++)
    {
   	 cin>>a[i];
    }
    
    bubble(a,n);
    
    cout<<"\n sorted array is=>";
    for(int i=0;i<n;i++)
    {
   	 cout<<a[i]<<endl;
    }


return 0;
}
```

Merge sort

The main focus of this code is the mergesort function, which implements the Merge Sort algorithm. Here's how the code works:

The mergesort function takes an array a, the starting index i, and the ending index j as input.

If i is less than j, it means there is more than one element in the current subarray. In this case, the function proceeds with the merge sort algorithm.

The variable mid is calculated as the middle index of the subarray.

The #pragma omp parallel sections directive is used to parallelize the recursive calls to mergesort for the left and right subarrays.

Within the parallel region, the left and right subarrays are sorted independently using recursive calls to mergesort.

After the parallel sections, the merge function is called to merge the sorted left and right subarrays.

The merge function takes two sorted subarrays (specified by their starting and ending indices) and merges them into a single sorted array. It uses an auxiliary array temp to store the merged result.

In the main function:

The user is prompted to enter the total number of elements and then the individual elements of the array.

The mergesort function is called with the array, starting index 0, and ending index n-1.

Finally, the sorted array is printed to the console.

OpenMP is used in this code to potentially parallelize the sorting process, allowing multiple threads to work on different portions of the array simultaneously. This can improve the sorting performance for large arrays. To compile and run this code with OpenMP support, you need to have a compiler that supports OpenMP and enable the OpenMP flag during compilation (e.g., -fopenmp for GCC).

```python
// Online C++ compiler to run C++ program online
#include<iostream>
#include<stdlib.h>
#include<omp.h>
using namespace std;


void mergesort(int a[],int i,int j);
void merge(int a[],int i1,int j1,int i2,int j2);

void mergesort(int a[],int i,int j)
{
	int mid;
	if(i<j)
	{
    	mid=(i+j)/2;
   	 
    	#pragma omp parallel sections
    	{

        	#pragma omp section
        	{
            	mergesort(a,i,mid);   	 
        	}

        	#pragma omp section
        	{
            	mergesort(a,mid+1,j);    
        	}
    	}

    	merge(a,i,mid,mid+1,j);    
	}

}
 
void merge(int a[],int i1,int j1,int i2,int j2)
{
	int temp[1000];    
	int i,j,k;
	i=i1;    
	j=i2;    
	k=0;
    
	while(i<=j1 && j<=j2)    
	{
    	if(a[i]<a[j])
    	{
        	temp[k++]=a[i++];
    	}
    	else
    	{
        	temp[k++]=a[j++];
    }    
	}
    
	while(i<=j1)    
	{
    	temp[k++]=a[i++];
	}
   	 
	while(j<=j2)    
	{
    	temp[k++]=a[j++];
	}
   	 
	for(i=i1,j=0;i<=j2;i++,j++)
	{
    	a[i]=temp[j];
	}    
}


int main()
{
	int *a,n,i;
	cout<<"\n enter total no of elements=>";
	cin>>n;
	a= new int[n];

	cout<<"\n enter elements=>";
	for(i=0;i<n;i++)
	{
    	cin>>a[i];
	}
   //	 start=.......
//#pragma omp…..
	mergesort(a, 0, n-1);
//          stop…….
	cout<<"\n sorted array is=>";
	for(i=0;i<n;i++)
	{
    	cout<<"\n"<<a[i];
	}
  	// Cout<<Stop-Start
	return 0;
}
```
